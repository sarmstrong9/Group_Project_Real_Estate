```python
# Importing dependencies
import pandas as pd 
import numpy as np
```

```python
# Loading data from redfin tsv and adding into dataframe
redFin_df = pd.read_csv("state_market_tracker.tsv000", sep = '\t', header = 0)
pd.set_option('display.max_columns', None)
redFin_df
```

```python
# Listing columns to determine what columns to keep
for col in redFin_df.columns:
    print(col)
```

```python
# Creating redfin dataframe with specified column
clean_redFin = redFin_df[['period_begin','state', 'property_type', 'median_sale_price', 'homes_sold' ]]
clean_redFin['year'] = pd.DatetimeIndex(redFin_df['period_begin']).year
clean_redFin = clean_redFin.drop(columns = ['period_begin'])
clean_redFin= clean_redFin.set_index('year')
clean_redFin
```

```python
# Creating California DataFrame
cali_df = clean_redFin[clean_redFin['state'] == 'California']
cali_df
```

```python
# Creating Texas DataFrame
texas_df = clean_redFin[clean_redFin['state'] == 'Texas']
texas_df
```

```python
# Adding migration DataFrame
migration_df = pd.read_csv("Migration_Flows_from_2010_to_2019.csv", header = 0)
migration_df
```

```python
#Selecting columns with data related to analysis 
migration_df_clean = migration_df[['current_state', 'year', 'population', 'from', 'number_of_people','from_different_state_Total']]
migration_df_clean = migration_df_clean.set_index('year')
migration_df_clean
```

```python
# Creating dataframe of migration to California from Texas
from_texas_to_cali = migration_df_clean.loc[(migration_df_clean['current_state'] == 'California')
                               & (migration_df_clean['from'] == 'Texas')]
from_texas_to_cali
```

```python
# Creating dataframe of migration from Texas to California
from_cali_to_texas = migration_df_clean.loc[(migration_df_clean['current_state'] == 'Texas')
                               & (migration_df_clean['from'] == 'California')]
from_cali_to_texas = from_cali_to_texas.reset_index()
from_cali_to_texas
```

```python
full_df = pd.concat([cali_df, texas_df, from_cali_to_texas])
full_df
```
