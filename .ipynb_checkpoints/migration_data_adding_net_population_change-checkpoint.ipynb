```python
import pandas as pd
migration_df = pd.read_csv('Resources/Migration_Flows_from_2010_to_2019.csv')
migration_df
```

```python
states = migration_df["current_state"].unique()
states
```

```python
state_data = migration_df.loc[migration_df["current_state"] == "Alabama"]
state_data
```

```python
state_year = state_data.drop_duplicates(subset=['population'], keep="last")
state_year.sort_values(by=["year"])
state_year
```

```python
population_change = state_year["population"]
population_change
```

```python
population_change = population_change.diff()
population_change
```

```python
population_list = population_change.to_list()
population_list
```

```python
population_list[0]
```

```python
state_data["population_change"] = ""
```

```python
state_data.sort_values(by=["year"])
```

```python
looped_year = 2010
for x in range(len(population_list)):
    state_data["population_change"].loc[(state_data["year"] == looped_year)] = population_list[x]
    print(looped_year)
    print(population_list[x])
    looped_year = looped_year + 1
state_data
```

```python
state_data.sort_values(by=["year"])
```

```python
state_data["population_change"] = ""
migration_df["population_change"] = ""

for state in states:
    state_data = migration_df.loc[migration_df["current_state"] == state]
    print(state)
    state_year = state_data.drop_duplicates(subset=['population'], keep="last")
    state_year.sort_values(by=["year"])
    population_change = state_year["population"]
    population_change = population_change.diff()
    population_list = population_change.to_list()
    print(population_list)
    state_data["population_change"] = ""
    
    looped_year = 2010
    for x in range(len(population_list)):
        state_data["population_change"].loc[(state_data["year"] == looped_year)] = population_list[x]
        print(looped_year)
        print(population_list[x])
        looped_year = looped_year + 1
    
    state_population_list = state_data["population_change"].to_list()

    migration_df["population_change"].loc[(migration_df["current_state"] == state)] = state_population_list
    
migration_df
```

```python
migration_df["population_change"].loc[(migration_df["year"] == 2010)] = 0
```

```python
migration_df = migration_df.sort_values(by = ["current_state", "year"])
```
